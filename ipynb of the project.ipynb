<h3 style="text-align: center;"><b>1 сценарий</b></h3>

В первом сценарии упор делается на разработку демо, в которое встривается нейросетевая модель для детекции.

Демо должно показывать, что детектор успешно отрабатывает на поданных изображениях и находит нужные объекты. Оформлять красиво визуальную часть демо -- приветствуется, но не обязательно -- это может быть просто кнопка "загрузить" и показ фото до/после детекции.
